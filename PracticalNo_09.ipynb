## 1. What is a Data Science pipeline? Explain its key stages with examples.

A **Data Science pipeline** is a structured series of steps that data scientists follow to transform raw data into actionable insights or predictions. It ensures a systematic approach to problem-solving and helps manage the complexity of data science projects.

**Key Stages with Examples:**

1.  **Problem Definition/Understanding:**
    *   **Description:** Clearly defining the business problem, objectives, and success metrics. This is crucial for guiding the entire process.
    *   **Example:** A retail company wants to predict customer churn to proactively offer incentives and retain valuable customers. The objective is to identify customers at high risk of churning within the next month.

2.  **Data Acquisition:**
    *   **Description:** Collecting raw data from various sources relevant to the defined problem.
    *   **Example:** Gathering customer transaction history, demographic information, website interaction logs, and customer service records from databases, APIs, and external files.

3.  **Data Cleaning/Preparation:**
    *   **Description:** Handling missing values, correcting errors, removing duplicates, and transforming data into a consistent format suitable for analysis.
    *   **Example:** Filling in missing age values using imputation techniques (e.g., mean, median), standardizing address formats, removing duplicate customer entries, and correcting typos in product categories.

4.  **Exploratory Data Analysis (EDA):**
    *   **Description:** Analyzing data to discover patterns, detect anomalies, test hypotheses, and extract key insights. Often involves visualizations and descriptive statistics.
    *   **Example:** Visualizing the distribution of customer age, analyzing the correlation between purchase frequency and churn, and identifying outliers in transaction amounts.

5.  **Feature Engineering:**
    *   **Description:** Creating new features or transforming existing ones to improve the performance of machine learning models.
    *   **Example:** Calculating a customer's 'average monthly spending,' 'number of days since last purchase,' or 'loyalty score' based on their transaction history.

6.  **Model Selection/Training:**
    *   **Description:** Choosing appropriate machine learning algorithms and training them on the prepared data.
    *   **Example:** Selecting a Logistic Regression, Random Forest, or Gradient Boosting model, splitting data into training and validation sets, and training the chosen model to predict churn.

7.  **Model Evaluation:**
    *   **Description:** Assessing the model's performance using relevant metrics and techniques (e.g., accuracy, precision, recall, F1-score, ROC curve).
    *   **Example:** Evaluating the churn prediction model's accuracy on unseen data, checking its precision and recall to ensure it correctly identifies churning customers without too many false positives.

8.  **Model Deployment/Monitoring:**
    *   **Description:** Integrating the trained model into a production environment and continuously monitoring its performance, retraining as needed.
    *   **Example:** Deploying the churn model as an API endpoint, integrating it with the CRM system to flag at-risk customers, and setting up alerts for performance degradation over time.

9.  **Communication/Reporting:**
    *   **Description:** Presenting findings, insights, and model results to stakeholders in a clear and understandable manner.
    *   **Example:** Creating dashboards, reports, or presentations to communicate the churn predictions, the factors contributing to churn, and the business impact of the model.

## 2. What is Exploratory Data Analysis (EDA)? How do univariate and bivariate analyses help in understanding the dataset?

**Exploratory Data Analysis (EDA)** is an approach to analyzing data sets to summarize their main characteristics, often with visual methods. It is used for seeing what the data can tell us beyond the formal modeling or hypothesis testing task. EDA helps in understanding the data's structure, identifying patterns, detecting anomalies, and testing assumptions.

### How Univariate Analysis Helps:

**Univariate analysis** examines one variable at a time. Its primary goal is to describe the distribution of a single variable.

*   **Understanding Central Tendency:** It helps to find the typical or central value of the data (e.g., mean, median, mode).
    *   **Example:** The average age of customers is 35 years, or the most frequent product purchased is 'Product X'.
*   **Understanding Dispersion/Spread:** It shows how spread out the data points are (e.g., range, variance, standard deviation, interquartile range).
    *   **Example:** Customer ages range from 18 to 70, or the standard deviation of income is \$15,000, indicating significant variability.
*   **Identifying Outliers:** Visualizations like box plots and histograms can quickly highlight unusual data points that might be errors or important anomalies.
    *   **Example:** A histogram of transaction amounts might show a few extremely large values, indicating high-value customers or potential data entry errors.
*   **Visualizing Distribution:** Histograms and frequency plots show the shape of the data's distribution (e.g., normal, skewed, uniform).
    *   **Example:** A histogram of customer ratings might reveal a left-skewed distribution, meaning most ratings are high.

### How Bivariate Analysis Helps:

**Bivariate analysis** examines the relationship between two variables. Its goal is to understand how two variables interact with each other.

*   **Identifying Relationships/Correlations:** It helps to detect patterns, trends, and associations between two variables.
    *   **Example:** A scatter plot between 'advertising spend' and 'sales' might show a positive linear relationship, suggesting that increased advertising leads to higher sales.
*   **Comparing Groups:** It allows for comparisons between different categories or groups.
    *   **Example:** A bar chart comparing the average sales across different product categories can show which categories are top performers.
*   **Understanding Impact:** It helps to see how one variable might influence or be associated with another.
    *   **Example:** A box plot showing 'income' by 'education level' can illustrate if higher education levels are associated with higher incomes.
*   **Detecting Interactions:** In some cases, it can reveal interactions or dependencies that are not apparent when looking at variables individually.
    *   **Example:** Analyzing churn rate against both 'customer tenure' and 'customer support interactions' might show that long-term customers who contact support frequently have a higher churn risk.
*   **Feature Selection Insights:** Bivariate analysis can help identify features that are strongly related to the target variable, aiding in feature selection for machine learning models.
    *   **Example:** A correlation matrix might show that 'number of customer service calls' has a strong positive correlation with 'churn likelihood'.

## 3. Differentiate between Label Encoding and One-Hot Encoding. When should each be used?

Both **Label Encoding** and **One-Hot Encoding** are techniques used to convert categorical data into numerical format, which is required by most machine learning algorithms. However, they differ in how they represent the categories and have different use cases.

### Label Encoding

*   **Description:** Assigns a unique integer (label) to each category in a categorical feature. For example, if a feature 'Color' has categories 'Red', 'Green', 'Blue', Label Encoding might transform them into 0, 1, 2 respectively.
*   **Output:** A single numerical column where each number represents a category.
*   **Implicit Order:** Introduces an artificial ordinal relationship (0 < 1 < 2) between categories. The model might incorrectly interpret these numerical differences as reflecting some kind of order or magnitude.

*   **When to Use:**
    *   **Ordinal Categorical Data:** When the categorical feature has an inherent order or ranking. For example, 'Education Level' (High School, Bachelor's, Master's, PhD) can be encoded as 0, 1, 2, 3 because there is a clear hierarchy.
    *   **High Cardinality with Tree-Based Models:** Sometimes used for high-cardinality nominal features (many unique categories) when using tree-based algorithms (like Decision Trees, Random Forests, Gradient Boosting Machines). Tree-based models can often handle the induced ordinality better than linear models.
    *   **Memory Efficiency:** When memory usage is a concern, as it only adds one column per feature.

### One-Hot Encoding

*   **Description:** Converts each category value into a new binary (0 or 1) feature column. If a feature 'Color' has categories 'Red', 'Green', 'Blue', One-Hot Encoding creates three new columns: 'Color_Red', 'Color_Green', 'Color_Blue'. For an observation that is 'Red', 'Color_Red' would be 1 and the others 0.
*   **Output:** Creates multiple new binary columns (N-1 or N columns, where N is the number of unique categories, to avoid multicollinearity sometimes N-1 is used). Each observation will have a 1 in one of these new columns and 0s in the rest.
*   **No Implicit Order:** Does not introduce any false ordinal relationships, as each category is treated as distinct.

*   **When to Use:**
    *   **Nominal Categorical Data:** When the categorical feature has no inherent order or ranking. For example, 'City' (New York, London, Paris) or 'Gender' (Male, Female, Other).
    *   **Linear Models and Neural Networks:** Essential for models that are sensitive to numerical relationships between features, such as Linear Regression, Logistic Regression, Support Vector Machines (SVMs), and Neural Networks. These models would misinterpret the arbitrary ordering introduced by Label Encoding.
    *   **Avoiding Misinterpretation:** To ensure that the model does not infer any magnitude or order between categories that don't exist in reality.

### Summary Table

| Feature             | Label Encoding                             | One-Hot Encoding                                |
| :------------------ | :----------------------------------------- | :---------------------------------------------- |
| **Output**          | Single numerical column                    | Multiple binary columns                         |
| **Order Assumed**   | Yes (0, 1, 2, ...)                         | No                                              |
| **Use Case**        | Ordinal data, tree-based models (sometimes) | Nominal data, linear models, neural networks    |
| **Memory Usage**    | Lower                                      | Higher (creates more columns)                   |
| **Risk of Misinterpretation** | High for nominal data                      | Low                                             |

## 4. Why is feature scaling required? Compare normalization and standardization.

### Why is Feature Scaling Required?

**Feature scaling** is a crucial data preprocessing step used to standardize the range of independent variables or features of the data. It's often necessary because many machine learning algorithms are sensitive to the scale of the input features.

Here are the main reasons why feature scaling is required:

1.  **Algorithm Performance:** Many algorithms, especially those that calculate distances between data points or rely on gradient descent, perform poorly or converge slowly if features have vastly different scales.
    *   **Examples:** K-Nearest Neighbors (KNN), Support Vector Machines (SVMs), Logistic Regression, Neural Networks, K-Means Clustering.
    *   **Distance-Based Algorithms:** If one feature has a range of 0-10,000 and another 0-1, the feature with the larger range will dominate the distance calculation, making the smaller-ranged feature almost irrelevant.
    *   **Gradient Descent Based Algorithms:** If features have different scales, the cost function will have an elongated shape, making gradient descent oscillate and take longer to find the minimum.

2.  **Equal Contribution:** Scaling ensures that all features contribute equally to the distance calculation or optimization process. Without scaling, features with larger values might be weighted more heavily by the model simply due to their magnitude.

3.  **Regularization:** Scaling can be important for regularization techniques (L1, L2 regularization) that penalize large coefficients. If features are not scaled, coefficients for features with smaller ranges might become disproportionately large to compensate, leading to unfair penalties.

4.  **Interpretability of Coefficients:** In models like Linear Regression, scaled features can make the interpretation of coefficients more meaningful, as a unit change in a scaled feature would have a comparable impact on the target.

### Comparison of Normalization and Standardization

**Normalization (Min-Max Scaling)**

*   **Method:** Rescales the feature values to a fixed range, typically between 0 and 1 (or -1 and 1). It subtracts the minimum value and divides by the range (max - min).
    *   **Formula:** `X_scaled = (X - X_min) / (X_max - X_min)`
*   **Effect:** Transforms data such that the minimum value is 0 and the maximum value is 1.
*   **Sensitivity to Outliers:** Highly sensitive to outliers. If there are extreme values, they will compress the majority of the data into a very small range, making it less effective.
*   **When to Use:**
    *   When the data distribution is *not* Gaussian (normal).
    *   When algorithms require inputs to be within a specific bounded range (e.g., some neural network activation functions expecting inputs between 0 and 1).
    *   Good for algorithms that do not assume a Gaussian distribution, like K-Nearest Neighbors (KNN) or some types of neural networks.

**Standardization (Z-score Normalization)**

*   **Method:** Rescales the feature values so that they have a mean of 0 and a standard deviation of 1. It subtracts the mean and divides by the standard deviation.
    *   **Formula:** `X_scaled = (X - μ) / σ` (where `μ` is the mean and `σ` is the standard deviation)
*   **Effect:** Transforms data to have a mean of 0 and unit variance (standard deviation of 1). The values are not bounded to a specific range, but most data points will fall between -3 and 3 for a normally distributed feature.
*   **Sensitivity to Outliers:** Less affected by outliers compared to normalization, as it uses the mean and standard deviation which are less sensitive than min/max values. However, outliers will still exist as larger deviations from the mean.
*   **When to Use:**
    *   When the data distribution is approximately Gaussian (normal).
    *   When algorithms assume a Gaussian distribution or are sensitive to feature scales, such as Linear Regression, Logistic Regression, SVMs, PCA, and algorithms that use gradient descent.
    *   When the range of data is unknown or varies significantly from observation to observation.

### Key Differences Summary

| Aspect                | Normalization (Min-Max Scaling)               | Standardization (Z-score Scaling)              |
| :-------------------- | :-------------------------------------------- | :--------------------------------------------- |
| **Formula**           | `(X - X_min) / (X_max - X_min)`               | `(X - μ) / σ`                                  |
| **Range**             | Fixed (e.g., [0, 1] or [-1, 1])               | Not bounded (typically around [-3, 3])         |
| **Effect on Mean/Std** | Does not guarantee mean=0, std=1              | Guarantees mean=0, std=1                       |
| **Outlier Sensitivity** | High                                          | Lower                                          |
| **Use Cases**         | Bounded ranges, non-Gaussian data, KNN, Neural Nets | Gaussian data, distance-based algos, LR, SVM, PCA |

In practice, it's often a good idea to try both and see which one yields better results for your specific model and dataset, or to use cross-validation to select the best scaling method.

## 5. What is feature engineering? Explain how creating new features can improve model performance.

### What is Feature Engineering?

**Feature engineering** is the process of using domain knowledge to extract or create new features from raw data to make machine learning algorithms work better. It's an art as much as a science, involving careful thought, experimentation, and understanding of both the data and the problem at hand. The goal is to transform the existing data into a format that best represents the underlying problem to the predictive model, thereby improving model accuracy and interpretability.

### How Creating New Features Can Improve Model Performance

Creating new features can significantly improve model performance by:

1.  **Making Hidden Patterns Explicit:** Raw data often contains information that is implicitly present but not directly usable by a model. Feature engineering makes these hidden patterns explicit.
    *   **Example:** From a `timestamp` column, you can extract `day_of_week`, `hour_of_day`, `month`, `is_weekend`, etc. A model might struggle to learn seasonality from a raw timestamp, but it can easily learn that sales are higher on weekends if `is_weekend` is a direct feature.

2.  **Capturing Non-Linear Relationships:** Simple linear models might fail to capture complex, non-linear relationships. Engineered features can transform these relationships into a more linear or digestible format for the model.
    *   **Example:** Instead of using raw `age`, creating `age_squared` or `age_group` (e.g., young, middle-aged, senior) can help models capture non-linear effects, such as very young and very old customers behaving similarly in some contexts.

3.  **Reducing Dimensionality and Noise (Implicitly):** Sometimes, a single engineered feature can summarize the information from multiple raw features, effectively reducing dimensionality and potentially noise.
    *   **Example:** Creating a `BMI` (Body Mass Index) feature from `height` and `weight` consolidates two pieces of information into one meaningful metric that directly relates to health categories.

4.  **Improving Model Interpretability:** Well-crafted features can make the model's decisions more understandable.
    *   **Example:** If predicting housing prices, a feature like `distance_to_city_center` is more interpretable than just raw latitude and longitude coordinates. A model's reliance on `distance_to_city_center` makes intuitive sense.

5.  **Incorporating Domain Knowledge:** Domain experts often have insights into what variables are truly important and how they interact. Feature engineering allows incorporating this valuable knowledge directly into the dataset.
    *   **Example:** In finance, calculating a `debt_to_income_ratio` from `debt` and `income` is a common practice because financial experts know this ratio is a strong indicator of creditworthiness.

6.  **Handling Categorical Data Effectively:** Transforming categorical variables into numerical ones (as discussed with one-hot encoding or label encoding) is a form of feature engineering that allows models to process them.
    *   **Example:** Converting `city_name` into `city_population` or `city_type` (urban, rural) can provide more meaningful numerical features than just one-hot encoding a long list of city names.

7.  **Addressing Missing Values:** Creating indicator features (e.g., `is_missing_age`) can sometimes be a form of feature engineering, informing the model when data was absent and allowing it to learn if missingness itself is predictive.

### Examples of Feature Engineering Techniques:

*   **Numerical Features:**
    *   **Binning:** Grouping continuous values into discrete bins (e.g., age into 0-18, 19-35, 36-60, 60+).
    *   **Interactions:** Multiplying or dividing two features (e.g., `price_per_square_foot`).
    *   **Polynomial Features:** Raising features to a power (e.g., `x^2`, `x^3`).
    *   **Log/Square Root Transformation:** Applying mathematical functions to normalize skewed distributions.
*   **Date/Time Features:**
    *   Extracting `year`, `month`, `day`, `day_of_week`, `hour`, `minute`.
    *   Calculating `time_since_event` or `time_until_event`.
    *   Creating `is_weekend`, `is_holiday`.
*   **Categorical Features:**
    *   **One-Hot Encoding, Label Encoding.**
    *   **Frequency Encoding:** Replacing categories with their frequency count in the dataset.
    *   **Target Encoding:** Replacing categories with the mean of the target variable for that category.
    *   **Combining Rare Categories:** Grouping infrequent categories into an 'Other' category.
*   **Text Features:**
    *   **Bag-of-Words, TF-IDF.**
    *   Extracting `word_count`, `character_count`, `average_word_length`.
    *   Sentiment scores.

By carefully engineering features, data scientists can provide models with a richer, more informative representation of the data, allowing them to uncover more complex relationships and make more accurate predictions.

```python
Code :

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns

print("### Data Science Pipeline Demonstration ###")

# --- 1. Data Loading (Synthetic Data) ---
print("\n--- 1. Data Loading ---")
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 28, 33, np.nan, 42, 38, 52, 29, 31, 37, 41, 48, 53],
    'Salary': [50000, 60000, 75000, 80000, 90000, 110000, 100000, 120000, 55000, 68000, 70000, 85000, 72000, 105000, 58000, 63000, 78000, 82000, 95000, 115000],
    'City': ['New York', 'London', 'Paris', 'New York', 'London', 'Paris', 'New York', 'London', 'Paris', 'New York', 'London', 'Paris', 'New York', 'London', 'Paris', 'New York', 'London', 'Paris', 'New York', 'London'],
    'Education_Level': ['High School', 'Bachelors', 'Masters', 'PhD', 'Bachelors', 'Masters', 'High School', 'PhD', 'Bachelors', 'Masters', 'High School', 'Bachelors', 'Masters', 'PhD', 'Bachelors', 'High School', 'Masters', 'Bachelors', 'PhD', 'Masters'],
    'Experience_Years': [2, 7, 10, 15, 12, 20, 18, 25, 5, 8, 3, 16, 11, 22, 6, 9, 14, 17, 19, 23],
    'Purchases': [5, 8, 12, 15, 10, 18, 14, 20, 7, 9, 6, 16, 11, 19, 8, 10, 13, 17, 15, 21],
    'Churn': [0, 0, 0, 1, 0, 1, 0, 1, 0, 0, 0, 1, 0, 1, 0, 0, 0, 1, 0, 1]
}
df = pd.DataFrame(data)
print("Original DataFrame head:")
display(df.head())
print("\nDataFrame Info:")
df.info()

# --- 2. Data Cleaning (Handling Missing Values) ---
print("\n--- 2. Data Cleaning ---")
# Impute missing 'Age' values with the mean
imputer = SimpleImputer(strategy='mean')
df['Age'] = imputer.fit_transform(df[['Age']])
print("Missing values after imputation:")
print(df.isnull().sum())

# --- 3. Exploratory Data Analysis (EDA) ---
print("\n--- 3. Exploratory Data Analysis (EDA) ---")

# Univariate Analysis (Distribution of Age)
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Bivariate Analysis (Salary vs. Experience_Years)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Experience_Years', y='Salary', hue='Churn', data=df)
plt.title('Salary vs. Experience Years (colored by Churn)')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.show()

# Bivariate Analysis (Average Salary by Education_Level)
plt.figure(figsize=(10, 5))
sns.boxplot(x='Education_Level', y='Salary', data=df, order=['High School', 'Bachelors', 'Masters', 'PhD'])
plt.title('Salary Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Salary')
plt.show()

# --- 4. Feature Engineering ---
print("\n--- 4. Feature Engineering ---")
# Create a new feature: 'Salary_Per_Year_Experience'
df['Salary_Per_Year_Experience'] = df['Salary'] / (df['Experience_Years'] + 1) # Add 1 to avoid division by zero
print("DataFrame with new feature 'Salary_Per_Year_Experience':")
display(df[['Salary', 'Experience_Years', 'Salary_Per_Year_Experience']].head())

# Create an 'Age_Group' feature using binning
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 45, 60, 100], labels=['Young', 'Middle-Aged', 'Senior', 'Elderly'])
print("DataFrame with new feature 'Age_Group':")
display(df[['Age', 'Age_Group']].head())

# --- 5. Feature Encoding ---
print("\n--- 5. Feature Encoding ---")

# Label Encoding (for ordinal 'Education_Level')
# Note: It's important to define the order for ordinal features
education_order = ['High School', 'Bachelors', 'Masters', 'PhD']
df['Education_Level_Encoded'] = df['Education_Level'].astype(pd.CategoricalDtype(categories=education_order, ordered=True)).cat.codes
print("DataFrame with 'Education_Level_Encoded':")
display(df[['Education_Level', 'Education_Level_Encoded']].head())

# One-Hot Encoding (for nominal 'City')
# Using ColumnTransformer for better pipeline integration and handling of remaining columns
ct = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['City'])
    ],
    remainder='passthrough' # Keep other columns
)

# Create a temporary DataFrame for one-hot encoding results
df_encoded_city = pd.DataFrame(ct.fit_transform(df), columns=ct.get_feature_names_out())
# Rename columns for clarity
df_encoded_city.columns = [col.replace('onehot__', '') for col in df_encoded_city.columns]

# Merge back with original df, dropping the original 'City' column to avoid redundancy
df_processed = df.drop(columns=['City']).reset_index(drop=True)
df_processed = pd.concat([df_processed, df_encoded_city.filter(like='City_')], axis=1)

print("DataFrame with One-Hot Encoded 'City' (first 5 rows and relevant columns):")
display(df_processed.filter(like='City_').head())
display(df_processed.head())

# --- 6. Feature Scaling ---
print("\n--- 6. Feature Scaling ---")

# Select numerical features for scaling (exclude encoded and target columns)
# Re-apply imputer as ColumnTransformer with remainder='passthrough' might convert to numpy array
numerical_cols = ['Age', 'Salary', 'Experience_Years', 'Purchases', 'Salary_Per_Year_Experience']

# Normalization (MinMaxScaler) on 'Salary'
min_max_scaler = MinMaxScaler()
df_processed['Salary_Normalized'] = min_max_scaler.fit_transform(df_processed[['Salary']])
print("DataFrame with 'Salary_Normalized':")
display(df_processed[['Salary', 'Salary_Normalized']].head())

# Standardization (StandardScaler) on 'Experience_Years'
standard_scaler = StandardScaler()
df_processed['Experience_Years_Standardized'] = standard_scaler.fit_transform(df_processed[['Experience_Years']])
print("DataFrame with 'Experience_Years_Standardized':")
display(df_processed[['Experience_Years', 'Experience_Years_Standardized']].head())

print("\n--- End of Pipeline Demonstration ---")
print("Final processed DataFrame head (showing key transformed features):")
display(df_processed[['Age', 'Salary_Normalized', 'Experience_Years_Standardized', 'Education_Level_Encoded', 'Age_Group', 'Salary_Per_Year_Experience'] + list(df_processed.filter(like='City_').columns)].head())
```

```python
print("\n--- 7. Model Training and Evaluation ---")

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Prepare features (X) and target (y)
# Drop original categorical columns and engineered features that are not directly numerical or encoded yet.
# Also drop 'Education_Level' and 'Age_Group' as we have their encoded versions.
features_to_drop = ['Education_Level', 'Age_Group', 'Churn']
X = df_processed.drop(columns=features_to_drop)
y = df_processed['Churn']

# Ensure all columns in X are numeric. Convert object columns if any remain after initial drops.
for col in X.columns:
    if X[col].dtype == 'object':
        try:
            X[col] = pd.to_numeric(X[col])
        except ValueError:
            print(f"Warning: Could not convert column '{col}' to numeric. Dropping it.")
            X = X.drop(columns=[col])

print("Features (X) head:")
display(X.head())
print("Target (y) head:")
display(y.head())

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"\nTraining set size: {len(X_train)} samples")
print(f"Test set size: {len(X_test)} samples")

# Initialize and train a Logistic Regression model
model = LogisticRegression(random_state=42, solver='liblinear') # 'liblinear' solver works well for small datasets
model.fit(X_train, y_train)

print("\nModel trained successfully!")

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"\nAccuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(report)
print("\nConfusion Matrix:")
print(conf_matrix)

print("\nThis step demonstrates how to use the preprocessed data to train a simple classification model and evaluate its performance.")
```
